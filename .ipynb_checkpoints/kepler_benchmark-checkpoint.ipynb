```python
# Kepler KOI Exoplanet Classification Benchmark
# Author: Andrii Mykuliak
# Dataset: NASA Kepler KOI Cumulative Table

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (accuracy_score, precision_score,
                             recall_score, f1_score, roc_auc_score,
                             confusion_matrix, roc_curve)
from xgboost import XGBClassifier

RANDOM_STATE = 42
print("Імпорти готові")
```

```python
# Завантаження датасету
df = pd.read_csv('cumulative.csv', comment='#')

print(f"Розмір датасету: {df.shape}")
print(f"\nРозподіл koi_disposition:")
print(df['koi_disposition'].value_counts())
```

```python
# Залишаємо тільки CONFIRMED і FALSE POSITIVE
df = df[df['koi_disposition'].isin(['CONFIRMED', 'FALSE POSITIVE'])]

# Features — тільки фізичні характеристики транзиту і зірки
# Виключаємо: ідентифікатори, похибки вимірювань, 
# діагностичні прапорці (koi_fpflag_*) і koi_score
features = [
    'koi_period',
    'koi_impact',
    'koi_duration',
    'koi_depth',
    'koi_prad',
    'koi_teq',
    'koi_model_snr',
    'koi_steff',
    'koi_slogg',
    'koi_srad',
    'koi_insol',
    'koi_kepmag'
]

target = 'koi_disposition'

# Вибираємо потрібні колонки і заповнюємо пропущені медіаною
df_model = df[features + [target]].copy()
df_model[features] = df_model[features].fillna(df_model[features].median())

# Кодуємо цільову змінну
df_model['target'] = (df_model[target] == 'CONFIRMED').astype(int)

print(f"Рядків після фільтрації: {df_model.shape[0]}")
print(f"Features: {len(features)}")
print(f"Пропущені значення: {df_model[features].isnull().sum().sum()}")
print(f"\nРозподіл класів:")
print(df_model[target].value_counts())
```

```python
X = df_model[features]
y = df_model['target']

# Train/test split — 80/20, зберігаємо пропорції класів
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y
)

# Scaling для лінійних моделей (LR, SVM)
# fit тільки на train щоб уникнути data leakage
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Train: {X_train.shape[0]} рядків")
print(f"Test:  {X_test.shape[0]} рядків")
print(f"\nРозподіл класів у train:")
print(y_train.value_counts(normalize=True).round(3) * 100)
```

```python
# Навчання моделей
models = {
    'Logistic Regression': LogisticRegression(random_state=RANDOM_STATE, max_iter=1000),
    'Decision Tree':       DecisionTreeClassifier(random_state=RANDOM_STATE),
    'Random Forest':       RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE),
    'XGBoost':             XGBClassifier(n_estimators=100, random_state=RANDOM_STATE, eval_metric='logloss'),
    'SVM':                 SVC(random_state=RANDOM_STATE, probability=True),
}

results = {}

for name, model in models.items():
    # LR і SVM потребують scaled даних
    if name in ['Logistic Regression', 'SVM']:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        y_prob = model.predict_proba(X_test_scaled)[:, 1]
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

    results[name] = {
        'Accuracy':  accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall':    recall_score(y_test, y_pred),
        'F1':        f1_score(y_test, y_pred),
        'ROC AUC':   roc_auc_score(y_test, y_prob),
        'y_pred':    y_pred,
        'y_prob':    y_prob,
    }
    print(f"✓ {name}")

print("\nВсі моделі навчені")
```

```python
metrics = ['Accuracy', 'Precision', 'Recall', 'F1', 'ROC AUC']

df_results = pd.DataFrame(
    {name: {m: results[name][m] for m in metrics} for name in results}
).T.round(4)

print("=== Benchmark Results ===\n")
print(df_results.to_string())
print("\n--- Найкращий результат по кожній метриці ---")
print(df_results.idxmax())
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for ax, name in zip(axes, results.keys()):
    cm = confusion_matrix(y_test, results[name]['y_pred'])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                xticklabels=['False Positive', 'Confirmed'],
                yticklabels=['False Positive', 'Confirmed'])
    ax.set_title(name, fontsize=10)
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

plt.suptitle('Confusion Matrices — Kepler KOI Classification', 
             fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig('confusion_matrices.png', dpi=150, bbox_inches='tight')
plt.show()
print("Збережено: confusion_matrices.png")
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

for name in results:
    fpr, tpr, _ = roc_curve(y_test, results[name]['y_prob'])
    auc = results[name]['ROC AUC']
    ax.plot(fpr, tpr, label=f'{name} (AUC={auc:.4f})')

ax.plot([0, 1], [0, 1], 'k--', label='Random baseline')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('ROC Curves — Kepler KOI Classification')
ax.legend(loc='lower right')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('roc_curves.png', dpi=150, bbox_inches='tight')
plt.show()
print("Збережено: roc_curves.png")
```

```python
rf_model = models['Random Forest']

importances = pd.Series(
    rf_model.feature_importances_,
    index=features
).sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(8, 5))
importances.plot(kind='barh', ax=ax, color='steelblue')
ax.set_title('Feature Importance — Random Forest')
ax.set_xlabel('Importance')
ax.grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.savefig('feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nТоп-5 найважливіших features:")
print(importances.sort_values(ascending=False).head(5).round(4))
```

```python
from sklearn.pipeline import Pipeline

seeds = [0, 7, 21, 42, 99]
stability_models = {
    'Logistic Regression': Pipeline([
        ('scaler', StandardScaler()),
        ('lr', LogisticRegression(max_iter=1000)),
    ]),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(n_estimators=100),
    'XGBoost':       XGBClassifier(n_estimators=100, eval_metric='logloss'),
    'SVM': Pipeline([
        ('scaler', StandardScaler()),
        ('svm', SVC(probability=True, kernel='rbf')),
    ]),
}

print("=== Stability across random seeds (ROC AUC, 5-fold CV) ===\n")

stability_results = {}

for name, model in stability_models.items():
    scores = []
    for seed in seeds:
        if name == 'Logistic Regression':
            model.set_params(lr__random_state=seed)
        elif name == 'SVM':
            model.set_params(svm__random_state=seed)
        else:
            model.set_params(random_state=seed)
        cv = cross_val_score(model, X, y, cv=5, scoring='roc_auc')
        scores.append(cv.mean())
    stability_results[name] = scores
    print(f"{name}:")
    print(f"  Scores: {[round(s, 4) for s in scores]}")
    print(f"  Mean:   {np.mean(scores):.4f}")
    print(f"  Std:    {np.std(scores):.4f}\n")
```

```python
stability_df = pd.DataFrame({
    name: {
        'Mean ROC AUC': round(np.mean(scores), 4),
        'Std':          round(np.std(scores), 4),
        'Min':          round(min(scores), 4),
        'Max':          round(max(scores), 4),
    }
    for name, scores in stability_results.items()
}).T

print("=== Stability across random seeds ===\n")
print(stability_df.to_string())
```

```python
# === Application to CANDIDATE class ===

# Відновлюємо кандидатів
df_original = pd.read_csv('cumulative.csv', comment='#')
df_candidates = df_original[df_original['koi_disposition'] == 'CANDIDATE'][features].copy()
df_candidates[features] = df_candidates[features].fillna(df_candidates[features].median())

print(f"Кандидатів для аналізу: {len(df_candidates)}\n")

# Scaled версія для LR і SVM
X_candidates = df_candidates[features]
X_candidates_scaled = scaler.transform(X_candidates)

# Передбачення всіх 5 моделей
df_candidates['prob_LR']  = models['Logistic Regression'].predict_proba(X_candidates_scaled)[:, 1]
df_candidates['prob_DT']  = models['Decision Tree'].predict_proba(X_candidates)[:, 1]
df_candidates['prob_RF']  = models['Random Forest'].predict_proba(X_candidates)[:, 1]
df_candidates['prob_XGB'] = models['XGBoost'].predict_proba(X_candidates)[:, 1]
df_candidates['prob_SVM'] = models['SVM'].predict_proba(X_candidates_scaled)[:, 1]

prob_cols = ['prob_LR', 'prob_DT', 'prob_RF', 'prob_XGB', 'prob_SVM']

# Бінарні передбачення (поріг 0.5)
for col in prob_cols:
    df_candidates[col.replace('prob_', 'pred_')] = (df_candidates[col] > 0.5).astype(int)

pred_cols = ['pred_LR', 'pred_DT', 'pred_RF', 'pred_XGB', 'pred_SVM']

# Кількість голосів "планета"
df_candidates['votes_planet'] = df_candidates[pred_cols].sum(axis=1)

# Консистентність
total = len(df_candidates)
all_agree_planet  = df_candidates[df_candidates['votes_planet'] == 5]
all_agree_not     = df_candidates[df_candidates['votes_planet'] == 0]
majority_planet   = df_candidates[df_candidates['votes_planet'] >= 3]
full_disagreement = df_candidates[df_candidates['votes_planet'].isin([2, 3])]

print("=== Consistency Analysis on Candidates (all 5 models) ===\n")
print(f"Всього кандидатів:                    {total}")
print(f"Всі 5 → планета:                      {len(all_agree_planet)} ({len(all_agree_planet)/total*100:.1f}%)")
print(f"Всі 5 → не планета:                   {len(all_agree_not)} ({len(all_agree_not)/total*100:.1f}%)")
print(f"Більшість (3+) → планета:             {len(majority_planet)} ({len(majority_planet)/total*100:.1f}%)")
print(f"Найбільша розбіжність (2 vs 3):       {len(full_disagreement)} ({len(full_disagreement)/total*100:.1f}%)")
print(f"\nПовна консистентність (0 або 5):      {(len(all_agree_planet)+len(all_agree_not))/total*100:.1f}%")

# Гістограма розподілу голосів
vote_dist = df_candidates['votes_planet'].value_counts().sort_index()

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(vote_dist.index, vote_dist.values, color='steelblue', edgecolor='white')

for bar, (votes, count) in zip(bars, vote_dist.items()):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
            f'{count}\n({count/total*100:.1f}%)',
            ha='center', va='bottom', fontsize=10)

ax.set_xlabel('Number of models voting "planet"')
ax.set_ylabel('Number of candidates')
ax.set_title('Model Voting Distribution on CANDIDATE class (N=1979)')
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels(['0\n(all: no)', '1', '2', '3', '4', '5\n(all: yes)'])
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('candidate_voting.png', dpi=300, bbox_inches='tight')
plt.show()
print("Збережено: candidate_voting.png")
```

```python
# Попарна консистентність між алгоритмами
import itertools

# Групуємо за типом моделі: спочатку дерева, потім лінійні/kernel
model_names_grouped = ['RF', 'XGB', 'DT', 'LR', 'SVM']
pred_cols_grouped   = ['pred_RF', 'pred_XGB', 'pred_DT', 'pred_LR', 'pred_SVM']

consistency_grouped = pd.DataFrame(index=model_names_grouped,
                                   columns=model_names_grouped, dtype=float)

for m1, c1 in zip(model_names_grouped, pred_cols_grouped):
    for m2, c2 in zip(model_names_grouped, pred_cols_grouped):
        agree = (df_candidates[c1] == df_candidates[c2]).sum()
        consistency_grouped.loc[m1, m2] = round(agree / total * 100, 1)

print("=== Pairwise Consistency between Models on Candidates ===\n")
print(consistency_grouped.to_string())

# Heatmap з кластерами
fig, ax = plt.subplots(figsize=(7, 6))
sns.heatmap(consistency_grouped.astype(float),
            annot=True, fmt='.1f', cmap='Blues',
            ax=ax, vmin=40, vmax=100,
            xticklabels=model_names_grouped,
            yticklabels=model_names_grouped)

# Обводимо кластер дерев (RF, XGB, DT) — лівий верхній 3x3
ax.add_patch(plt.Rectangle((0, 0), 3, 3, fill=False,
                             edgecolor='tomato', linewidth=3, clip_on=False))

# Обводимо кластер лінійних (LR, SVM) — правий нижній 2x2
ax.add_patch(plt.Rectangle((3, 3), 2, 2, fill=False,
                             edgecolor='green', linewidth=3, clip_on=False))

# Легенда
from matplotlib.patches import Patch
legend = [
    Patch(edgecolor='tomato', facecolor='none', linewidth=2,
          label='Tree-based cluster (RF, XGB, DT)'),
    Patch(edgecolor='green',  facecolor='none', linewidth=2,
          label='Linear/Kernel cluster (LR, SVM)'),
]
ax.legend(handles=legend, loc='upper right',
          bbox_to_anchor=(1.0, -0.12), fontsize=9)

ax.set_title('Pairwise Model Consistency on CANDIDATE class (%)\nGrouped by decision boundary type')
plt.tight_layout()
plt.savefig('pairwise_consistency_grouped.png', dpi=300, bbox_inches='tight')
plt.show()
print("Збережено: pairwise_consistency_grouped.png")
```

```python
# Розбиваємо на зони
grey      = df_candidates[df_candidates['votes_planet'].isin([2, 3])]
all_yes   = df_candidates[df_candidates['votes_planet'] == 5]
all_no    = df_candidates[df_candidates['votes_planet'] == 0]

# Порівняння середніх по ключових параметрах
key_features = ['koi_prad', 'koi_period', 'koi_model_snr', 
                'koi_impact', 'koi_depth', 'koi_duration']

comparison = pd.DataFrame({
    'All YES (5/5)': all_yes[key_features].mean(),
    'Grey Zone (2-3/5)': grey[key_features].mean(),
    'All NO (0/5)': all_no[key_features].mean(),
}).round(3)

print("=== Physical Parameters by Voting Zone ===\n")
print(comparison.to_string())
```

```python
# Додаємо зону до всіх кандидатів
df_candidates['zone'] = df_candidates['votes_planet'].map({
    0: 'All NO',
    1: 'Grey',
    2: 'Grey',
    3: 'Grey',
    4: 'Majority YES',
    5: 'All YES'
})

# Порівняння середніх по ключових параметрах
key_features = ['koi_prad', 'koi_period', 'koi_model_snr',
                'koi_impact', 'koi_depth', 'koi_duration']

all_yes = df_candidates[df_candidates['votes_planet'] == 5]
all_no  = df_candidates[df_candidates['votes_planet'] == 0]
grey    = df_candidates[df_candidates['zone'] == 'Grey']

comparison = pd.DataFrame({
    'All YES (5/5)':     all_yes[key_features].mean(),
    'Grey Zone (2-3/5)': grey[key_features].mean(),
    'All NO (0/5)':      all_no[key_features].mean(),
}).round(3)

print("=== Physical Parameters by Voting Zone ===\n")
print(comparison.to_string())

# Scatter plot
snr_max  = df_candidates['koi_model_snr'].quantile(0.95)
prad_max = df_candidates['koi_prad'].quantile(0.95)
per_max  = df_candidates['koi_period'].quantile(0.95)

colors = {
    'All NO':       'steelblue',
    'Grey':         'orange',
    'Majority YES': 'lightgreen',
    'All YES':      'green'
}

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for zone, group in df_candidates.groupby('zone'):
    axes[0].scatter(group['koi_model_snr'], group['koi_prad'],
                    c=colors[zone], label=zone, alpha=0.4, s=15)
    axes[1].scatter(group['koi_period'], group['koi_prad'],
                    c=colors[zone], label=zone, alpha=0.4, s=15)

axes[0].set_xlim(0, snr_max)
axes[0].set_ylim(0, prad_max)
axes[1].set_xlim(0, per_max)
axes[1].set_ylim(0, prad_max)

axes[0].set_xlabel('koi_model_snr (Signal-to-Noise)')
axes[0].set_ylabel('koi_prad (Planet Radius, Earth radii)')
axes[0].set_title('SNR vs Planet Radius')
axes[0].legend(fontsize=8)
axes[0].grid(True, alpha=0.3)

axes[1].set_xlabel('koi_period (Orbital Period, days)')
axes[1].set_ylabel('koi_prad (Planet Radius, Earth radii)')
axes[1].set_title('Period vs Planet Radius')
axes[1].legend(fontsize=8)
axes[1].grid(True, alpha=0.3)

plt.suptitle('Candidate Zones by Model Agreement', fontsize=13)
plt.tight_layout()
plt.savefig('candidate_zones_scatter.png', dpi=150, bbox_inches='tight')
plt.show()
print("Збережено: candidate_zones_scatter.png")
```

```python
# Hexbin density map — Grey Zone vs All YES (порівняння на одному графіку)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Фільтруємо обидві осі разом
ay = all_yes[(all_yes['koi_model_snr'] < snr_max) & (all_yes['koi_prad'] < prad_max)]
gy = grey[(grey['koi_model_snr'] < snr_max) & (grey['koi_prad'] < prad_max)]
ay2 = all_yes[(all_yes['koi_period'] < per_max) & (all_yes['koi_prad'] < prad_max)]
gy2 = grey[(grey['koi_period'] < per_max) & (grey['koi_prad'] < prad_max)]

# Лівий — SNR vs Radius
axes[0].hexbin(ay['koi_model_snr'], ay['koi_prad'],
               gridsize=25, cmap='Greens', mincnt=1, alpha=0.7)
axes[0].hexbin(gy['koi_model_snr'], gy['koi_prad'],
               gridsize=25, cmap='Oranges', mincnt=1, alpha=0.7)
axes[0].set_xlim(0, snr_max)
axes[0].set_ylim(0, prad_max)
axes[0].set_xlabel('koi_model_snr (Signal-to-Noise)')
axes[0].set_ylabel('koi_prad (Planet Radius, Earth radii)')
axes[0].set_title('SNR vs Planet Radius')
axes[0].grid(True, alpha=0.3)

# Правий — Period vs Radius
axes[1].hexbin(ay2['koi_period'], ay2['koi_prad'],
               gridsize=25, cmap='Greens', mincnt=1, alpha=0.7)
axes[1].hexbin(gy2['koi_period'], gy2['koi_prad'],
               gridsize=25, cmap='Oranges', mincnt=1, alpha=0.7)
axes[1].set_xlim(0, per_max)
axes[1].set_ylim(0, prad_max)
axes[1].set_xlabel('koi_period (Orbital Period, days)')
axes[1].set_ylabel('koi_prad (Planet Radius, Earth radii)')
axes[1].set_title('Period vs Planet Radius')
axes[1].grid(True, alpha=0.3)

# Легенда
from matplotlib.patches import Patch
legend = [
    Patch(facecolor='green',  alpha=0.7, label='All YES (311)'),
    Patch(facecolor='orange', alpha=0.7, label='Grey Zone (698)'),
]
for ax in axes:
    ax.legend(handles=legend, fontsize=9)

plt.suptitle('All YES vs Grey Zone — Density Comparison', fontsize=13)
plt.tight_layout()
plt.savefig('grey_zone_hexbin.png', dpi=300, bbox_inches='tight')
plt.show()
print("Збережено: grey_zone_hexbin.png")
```
